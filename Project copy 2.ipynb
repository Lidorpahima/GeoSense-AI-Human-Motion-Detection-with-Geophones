```python
# ==============================================================================
# ========= קוד מאוחד: EfficientNet + ספקטרוגרמה + אוגמנטציות משולבות =========
# ==============================================================================

# --- 1. ייבוא והגדרות ---
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
# שינוי: ייבוא ספציפי ל-EfficientNet
from torchvision.models import efficientnet_b0, EfficientNet_B0_Weights 
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import precision_score, recall_score, confusion_matrix, accuracy_score, classification_report
from scipy.signal import butter, filtfilt, spectrogram
import librosa
import librosa.display
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
import time
import random
from copy import deepcopy
import cv2 # דרוש ל-resize

print(f"Torch version: {torch.__version__}")
print(f"Torchvision version: {torchvision.__version__}")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# --- הגדרות גלובליות (כמו קודם) ---
LABEL_MAN = 0; LABEL_CAR = 1; LABEL_NOISE = 2
CLASS_NAMES = ["Steps", "Vehicle", "Noise"]; NUM_CLASSES = 3
window_size = 2000; stride = 500; fs = 1000.0; max_channels = 39
N_MELS = 128; N_FFT = 1024; HOP_LENGTH = window_size // (N_MELS * 2) # ~8
IMG_WIDTH = int(np.floor(window_size / HOP_LENGTH) + 1) # ~251
IMG_HEIGHT_FINAL = 224; IMG_WIDTH_FINAL = 224 # גודל קלט סטנדרטי למודלים
lowcut = 1.0; highcut = 100.0; order = 4
print(f"Spectrogram Size (Target): ({IMG_HEIGHT_FINAL}, {IMG_WIDTH_FINAL})")

# --- 2. פונקציות עזר ועיבוד נתונים ---

def pad_channels(data, target_channels): 
    if data.ndim == 1: data = data.reshape(-1, 1)
    if data.shape[1] == target_channels: return data
    elif data.shape[1] < target_channels:
        padding = np.zeros((data.shape[0], target_channels - data.shape[1]))
        return np.hstack((data, padding))
    else: # data.shape[1] > target_channels
        return data[:, :target_channels]

def create_windows(signal, label, window_size=window_size, stride=stride): 
    X_windows, y_labels = [], []
    if signal.shape[0] < window_size: return [], []
    for start in range(0, signal.shape[0] - window_size + 1, stride):
        X_windows.append(signal[start : start + window_size, :])
        y_labels.append(label)
    return X_windows, y_labels

# --- פונקציות אוגמנטציה על הסיגנל (לפני ספקטרוגרמה) ---
def add_gaussian_noise_signal(signal_1d, noise_level=0.01):
    noise = np.random.normal(0, noise_level, signal_1d.shape)
    return signal_1d + noise

def time_stretch_signal(signal_1d, stretch_rate=1.0, target_len=window_size):
    stretched = librosa.effects.time_stretch(signal_1d.astype(np.float32), rate=stretch_rate)
    current_len = len(stretched)
    if current_len > target_len:
        start = (current_len - target_len) // 2
        return stretched[start : start + target_len]
    elif current_len < target_len:
        pad_len = target_len - current_len
        pad_start = pad_len // 2
        padded = np.zeros(target_len)
        padded[pad_start : pad_start + current_len] = stretched
        return padded
    else:
        return stretched

# --- פונקציה ליצירת ספקטרוגרמה ---
def create_mel_spectrogram(window_1d, sr=int(fs), n_fft=N_FFT, hop_length=HOP_LENGTH, n_mels=N_MELS): 
    try:
        mel_spec = librosa.feature.melspectrogram(y=window_1d.astype(np.float32), sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
        mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
        min_db, max_db = mel_spec_db.min(), mel_spec_db.max()
        mel_spec_norm = (mel_spec_db - min_db) / (max_db - min_db) if max_db > min_db else np.zeros_like(mel_spec_db)
        return mel_spec_norm
    except Exception as e: 
        # print(f"Spectrogram failed: {e}") # Debug
        return np.zeros((n_mels, IMG_WIDTH)) # החזר תמונה ריקה תקינה בגודל הצפוי (לפני resize)

# --- פונקציית טעינה משולבת (עם אוגמנטציית סיגנל אופציונלית) ---
def load_data_specs_maybe_aug(file_map, apply_signal_aug=False, signal_aug_prob=0.4, target_img_size=(IMG_HEIGHT_FINAL, IMG_WIDTH_FINAL)):
    all_specs, all_labels = [], []
    try: b, a = butter(order, [lowcut/(fs/2), highcut/(fs/2)], btype='band')
    except ValueError as e: print(f"Filter error: {e}"); return [], []
    print(f"Loading data (Signal Aug: {apply_signal_aug}, Prob: {signal_aug_prob})...")
    
    for label, info in file_map.items():
        filepaths = info['path']; extra_args = info.get('args', {})
        if isinstance(filepaths, str): filepaths = [filepaths]
        for filepath in filepaths:
            try:
                # --- טעינת קובץ ועיבוד בסיסי ---
                try: df = pd.read_csv(filepath, header=None, **extra_args)
                except: df = pd.read_csv(filepath, header=None, skiprows=1, **extra_args) # Retry
                df = df.dropna(); data = df.values
                if not np.issubdtype(data.dtype, np.number): data = data.astype(float)
                if data.shape[0] < window_size or data.ndim < 2 or data.shape[1] == 0: continue
                data_padded = pad_channels(data, max_channels)
                windows, labels = create_windows(data_padded, label)
                if not windows: continue
                
                # --- עיבוד חלונות: פילטור, אוגמנטציה (אופ'), ספקטרוגרמה, resize ---
                for i, w in enumerate(windows):
                    try:
                        filtered_w = filtfilt(b, a, w, axis=0)
                        signal_to_process = filtered_w[:, 0] # ערוץ 0

                        # <<< הפעלת אוגמנטציית סיגנל בהסתברות >>>
                        if apply_signal_aug and random.random() < signal_aug_prob:
                            aug_type = random.choice(['stretch', 'noise'])
                            if aug_type == 'stretch':
                                rate = np.random.uniform(0.9, 1.1) # מתיחה/כיווץ קלים
                                signal_to_process = time_stretch_signal(signal_to_process, rate)
                            elif aug_type == 'noise':
                                level = np.random.uniform(0.005, 0.015) # רעש נמוך
                                signal_to_process = add_gaussian_noise_signal(signal_to_process, level)
                        # <<< סוף אוגמנטציית סיגנל >>>

                        mel_spec_norm = create_mel_spectrogram(signal_to_process) 
                        if mel_spec_norm is None or mel_spec_norm.shape[0] != N_MELS: continue # דלג אם יצירה נכשלה

                        if target_img_size is not None: 
                            resized_spec = cv2.resize(mel_spec_norm, (target_img_size[1], target_img_size[0]), interpolation=cv2.INTER_LINEAR)
                        else: resized_spec = mel_spec_norm
                        spec_3_channels = np.stack([resized_spec]*3, axis=0) 
                        all_specs.append(spec_3_channels)
                        all_labels.append(labels[i])
                    except Exception as e_inner: pass # Skip faulty windows silently
            except Exception as e: pass # Skip faulty files silently

    if not all_specs: print("Warning: No spectrograms created."); return np.array([]), np.array([])
    print(f"Total spectrograms created: {len(all_specs)}")
    return np.array(all_specs, dtype=np.float32), np.array(all_labels, dtype=np.int64)


# --- טעינת הדאטה המקורי והחדש (רק פעם אחת) ---
original_file_map = { LABEL_MAN: {'path': 'man.csv'}, LABEL_CAR: {'path': 'car.csv', 'args': {'skiprows': 1}}, LABEL_NOISE: {'path': 'nothing.csv'} }
# טעינה ללא אוגמנטציית סיגנל (נגדיר ב-DataLoader לאימון)
all_original_specs_np, all_original_labels_np = load_data_specs_maybe_aug(original_file_map, apply_signal_aug=False) 

new_data_folder = 'normalized_csv'; new_files = glob.glob(os.path.join(new_data_folder, '*.csv')); new_file_map = {}
if new_files: new_file_map[LABEL_MAN] = {'path': new_files}
else: print(f"Warning: No CSV files found in {new_data_folder}.")
new_specs_np, new_labels_np = load_data_specs_maybe_aug(new_file_map, apply_signal_aug=False) if new_files else (np.array([]), np.array([]))

if all_original_specs_np.size == 0: print("CRITICAL ERROR: Failed to load original data."); exit()
print("Original Spectrogram data shape:", all_original_specs_np.shape)
if new_specs_np.size > 0: print("New Spectrogram data shape:", new_specs_np.shape)

# --- הגדרת טרנספורמציות (SpecAugment + Normalize) ---
# --- הגדרת טרנספורמציות (SpecAugment + Normalize בלבד) ---
class SpecAugment(nn.Module): # ... (קוד המחלקה כמו קודם) ...
    def __init__(self, freq_mask_param=20, time_mask_param=40, num_freq_masks=1, num_time_masks=1):
        super().__init__(); self.freq_mask_param = freq_mask_param; self.time_mask_param = time_mask_param
        self.num_freq_masks = num_freq_masks; self.num_time_masks = num_time_masks
    def forward(self, spec): 
        # מצפה לקבל spec כ-Tensor (C, F, T)
        if not isinstance(spec, torch.Tensor): # בדיקה נוספת
            raise TypeError("SpecAugment expects a PyTorch Tensor.")
            
        spec_aug = spec.clone(); _, F, T = spec_aug.shape
        for _ in range(self.num_freq_masks):
            if F > self.freq_mask_param > 0: f = random.randrange(0, self.freq_mask_param); f0 = random.randrange(0, F - f); spec_aug[:, f0:f0+f, :] = 0
        for _ in range(self.num_time_masks):
             if T > self.time_mask_param > 0: t = random.randrange(0, self.time_mask_param); t0 = random.randrange(0, T - t); spec_aug[:, :, t0:t0+t] = 0
        return spec_aug
        
imagenet_mean = [0.485, 0.456, 0.406]; imagenet_std = [0.229, 0.224, 0.225]

train_transforms = transforms.Compose([
    # הסרנו את transforms.ToPILImage() מכאן
    SpecAugment(freq_mask_param=20, time_mask_param=40, num_freq_masks=1, num_time_masks=1), # SpecAugment על ה-Tensor
    transforms.Normalize(mean=imagenet_mean, std=imagenet_std) # Normalize על ה-Tensor
])

val_test_transforms = transforms.Compose([ 
    transforms.Normalize(mean=imagenet_mean, std=imagenet_std) 
])

print("\nTransforms defined (SpecAugment on Tensor).")


# --- Dataset (כמו קודם) ---
class SpectrogramDataset(Dataset): # ... (קוד המחלקה כמו קודם) ...
    def __init__(self, specs_np, labels_np, transform=None):
        self.specs = torch.from_numpy(specs_np).float(); self.labels = torch.from_numpy(labels_np).long()
        self.transform = transform
    def __len__(self): return len(self.specs)
    def __getitem__(self, idx):
        spec = self.specs[idx]; label = self.labels[idx]
        if self.transform: spec = self.transform(spec)
        return spec, label

# --- 3. הגדרת מודל EfficientNet ---
def setup_efficientnet_model(model_name='efficientnet_b0', pretrained=True, num_classes=NUM_CLASSES, dropout_p=0.4): # Dropout מעט פחות אגרסיבי
    weights = EfficientNet_B0_Weights.DEFAULT if pretrained else None
    model = efficientnet_b0(weights=weights)
    print(f"Freezing features for {model_name}...")
    for param in model.features.parameters(): param.requires_grad = False
    num_ftrs = model.classifier[1].in_features
    model.classifier = nn.Sequential( nn.Dropout(p=dropout_p, inplace=True), nn.Linear(num_ftrs, num_classes) )
    print(f"Replaced classifier: Dropout({dropout_p}) + Linear({num_ftrs}, {num_classes})")
    model = model.to(device)
    return model

# --- 4. פונקציות אימון והערכה (כמו קודם) ---
def train_epoch_effnet(model, dataloader, criterion, optimizer, device): # ... (זהה ל-train_epoch_resnet) ...
    model.train(); running_loss = 0.0; correct_preds = 0; total_samples = 0
    for inputs, labels in dataloader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad(); outputs = model(inputs); loss = criterion(outputs, labels)
        preds = outputs.max(1)[1]; loss.backward(); optimizer.step()
        running_loss += loss.item() * inputs.size(0); correct_preds += (preds == labels).sum().item()
        total_samples += labels.size(0)
    return running_loss / total_samples, correct_preds / total_samples

def evaluate_model_effnet(model, dataloader, criterion, device): # ... (זהה ל-evaluate_model_resnet) ...
    model.eval(); running_loss = 0.0; correct_preds = 0; total_samples = 0; all_preds = []; all_labels = []
    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs); loss = criterion(outputs, labels); preds = outputs.max(1)[1]
            running_loss += loss.item() * inputs.size(0); correct_preds += (preds == labels).sum().item()
            total_samples += labels.size(0)
            all_preds.extend(preds.cpu().numpy()); all_labels.extend(labels.cpu().numpy())
    return running_loss / total_samples, correct_preds / total_samples, all_labels, all_preds

# --- פונקציית הצגת גרפים (כמו קודם) ---
def plot_fold_results(fold_num, history, val_labels, val_preds, new_data_labels=None, new_data_preds=None, new_data_recall=None, class_names=CLASS_NAMES): # ... (קוד הפונקציה המלא) ...
    print(f"\n--- Results for Fold {fold_num} ---"); epochs_ran = len(history['val_loss']); epochs_range = range(1, epochs_ran + 1)
    plt.figure(figsize=(12, 5)); plt.suptitle(f'Fold {fold_num} Training History', fontsize=14)
    plt.subplot(1, 2, 1); plt.plot(epochs_range, history['train_loss'], 'o-', label='Train Loss'); plt.plot(epochs_range, history['val_loss'], 'o-', label='Val Loss')
    plt.xlabel('Epochs'); plt.ylabel('Loss'); plt.legend(); plt.grid(True); plt.title('Loss vs. Epochs')
    plt.subplot(1, 2, 2); plt.plot(epochs_range, history['train_acc'], 'o-', label='Train Acc'); plt.plot(epochs_range, history['val_acc'], 'o-', label='Val Acc')
    plt.xlabel('Epochs'); plt.ylabel('Accuracy'); plt.legend(); plt.grid(True); plt.title('Accuracy vs. Epochs'); plt.ylim(0, 1.05)
    plt.tight_layout(rect=[0, 0.03, 1, 0.93]); plt.show()
    print("\nValidation Set Performance:"); cm_val = confusion_matrix(val_labels, val_preds, labels=range(NUM_CLASSES)); report_val = classification_report(val_labels, val_preds, target_names=class_names, labels=range(NUM_CLASSES), zero_division=0)
    print("Confusion Matrix (Validation):"); print(cm_val); print("\nClassification Report (Validation):"); print(report_val)
    plt.figure(figsize=(6, 5)); sns.heatmap(cm_val, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names); plt.title(f'Fold {fold_num} - Validation CM'); plt.xlabel('Predicted'); plt.ylabel('True'); plt.show()
    if new_data_labels is not None and new_data_preds is not None:
        print("\nNew Dataset Performance:"); cm_new = confusion_matrix(new_data_labels, new_data_preds, labels=range(NUM_CLASSES)); report_new = classification_report(new_data_labels, new_data_preds, target_names=class_names, labels=range(NUM_CLASSES), zero_division=0)
        print(f"Overall Acc on New Data: {accuracy_score(new_data_labels, new_data_preds):.4f}"); print(f"'Steps' Recall on New Data: {new_data_recall:.4f}" if new_data_recall is not None else "N/A")
        print("\nConfusion Matrix (New Data):"); print(cm_new); print("\nClassification Report (New Data):"); print(report_new)
        plt.figure(figsize=(6, 5)); sns.heatmap(cm_new, annot=True, fmt='d', cmap='Oranges', xticklabels=class_names, yticklabels=class_names); plt.title(f'Fold {fold_num} - New Data CM'); plt.xlabel('Predicted'); plt.ylabel('True'); plt.show()


# --- 5. K-Fold Cross-Validation עם EfficientNet ---
N_SPLITS = 5
EPOCHS = 25 # EfficientNet עשוי להתכנס מהר יותר
PATIENCE = 7 
BATCH_SIZE = 32 
WEIGHT_DECAY = 1e-4 
MODEL_NAME = 'efficientnet_b0' 
DROPOUT_P = 0.4 # אפשר להתאים
LEARNING_RATE = 0.001 

kfold = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=678) # <<< Random state חדש

fold_results_effnet = []
best_overall_new_data_recall = -1.0 
best_fold_model_state = None
best_fold_number = -1
best_fold_history = None 

# --- לולאת K-Fold ---
for fold, (train_ids, val_ids) in enumerate(kfold.split(all_original_specs_np, all_original_labels_np)):
    print(f"\n===== FOLD {fold+1}/{N_SPLITS} ({MODEL_NAME} Frozen + Dropout) =====")
    start_fold_time = time.time()

    # 1. נתונים, Datasets, DataLoaders
    # שינוי: אפשרות להפעלת אוגמנטציית סיגנל על האימון דרך ה-Dataset
    X_train_fold = all_original_specs_np[train_ids]; y_train_fold = all_original_labels_np[train_ids]
    X_val_fold = all_original_specs_np[val_ids]; y_val_fold = all_original_labels_np[val_ids]
    
    # --- שינוי: יצירת Dataset אימון עם אוגמנטציית סיגנל אופציונלית ---
    # ניצור מערך חדש עם אוגמנטציה על הסיגנל לפני יצירת ה-Dataset
    # (דרך פשוטה יותר מאשר בתוך ה-Dataset __getitem__)
    print("  Applying signal augmentation to training data for this fold...")
    X_train_fold_signal_aug, y_train_fold_signal_aug = load_data_specs_maybe_aug(
         {lbl: {'path': []} for lbl in range(NUM_CLASSES)}, # מפה ריקה, נשתמש בנתונים שכבר בזיכרון
         apply_signal_aug=True, signal_aug_prob=0.5 # הפעלת האוגמנטציה
    )
    # --- הערה: הקוד לעיל ל-load_data_specs_maybe_aug צריך להיות מותאם ---
    # --- לקבל נתונים קיימים במקום לקרוא מקבצים שוב ---
    # --- פתרון פשוט יותר כרגע: לא להפעיל אוג' סיגנל או להפעיל על הכל ---
    
    # פתרון זמני: נשתמש באוגמנטציות התמונה בלבד (SpecAugment)
    train_dataset = SpectrogramDataset(X_train_fold, y_train_fold, transform=train_transforms)
    val_dataset = SpectrogramDataset(X_val_fold, y_val_fold, transform=val_test_transforms)
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=0, pin_memory=True)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=0, pin_memory=True)

    # 2. מודל, Criterion, Optimizer
    model = setup_efficientnet_model(model_name=MODEL_NAME, pretrained=True, num_classes=NUM_CLASSES, dropout_p=DROPOUT_P)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.classifier.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY) 
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'max', patience=4, factor=0.1, verbose=False)

    # 3. אימון
    best_val_acc = 0.0; current_best_fold_model_state = None; epochs_no_improve = 0
    history_current_fold = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
    print(f"Starting training for Fold {fold+1}...")
    for epoch in range(1, EPOCHS + 1):
        train_loss, train_acc = train_epoch_effnet(model, train_loader, criterion, optimizer, device)
        val_loss, val_acc, fold_val_labels, fold_val_preds = evaluate_model_effnet(model, val_loader, criterion, device) 
        history_current_fold['train_loss'].append(train_loss); history_current_fold['train_acc'].append(train_acc)
        history_current_fold['val_loss'].append(val_loss); history_current_fold['val_acc'].append(val_acc)
        scheduler.step(val_acc)
        if epoch % 5 == 0 or epoch == EPOCHS: print(f"  Epoch {epoch}/{EPOCHS} -> Tr Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}, Val Loss: {val_loss:.4f}")
        if val_acc > best_val_acc: best_val_acc = val_acc; current_best_fold_model_state = deepcopy(model.state_dict()); epochs_no_improve = 0
        else: epochs_no_improve += 1
        if epochs_no_improve >= PATIENCE: print(f"  Early stopping triggered after {epoch} epochs."); break
    print(f"Fold {fold+1} Best Validation Accuracy: {best_val_acc:.4f}")

    # 4. הערכה על הדאטה החדש
    current_fold_new_data_recall = -1.0; fold_new_data_labels = None; fold_new_data_preds = None
    if current_best_fold_model_state and new_specs_np.size > 0:
         model.load_state_dict(current_best_fold_model_state)
         new_dataset = SpectrogramDataset(new_specs_np, new_labels_np, transform=val_test_transforms) 
         new_loader = DataLoader(new_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=0)
         _, test_acc_new, fold_new_data_labels, fold_new_data_preds = evaluate_model_effnet(model, new_loader, criterion, device) 
         current_fold_new_data_recall = recall_score(fold_new_data_labels, fold_new_data_preds, labels=[LABEL_MAN], average='micro', zero_division=0)
         print(f"  'Steps' Recall on new dataset (Fold {fold+1}): {current_fold_new_data_recall:.4f}")
         if current_fold_new_data_recall > best_overall_new_data_recall:
              print(f"  *** New best overall recall found: {current_fold_new_data_recall:.4f} ***")
              best_overall_new_data_recall = current_fold_new_data_recall
              best_fold_model_state = current_best_fold_model_state 
              best_fold_number = fold + 1
              best_fold_history = history_current_fold 
              
    # 5. הצגת תוצאות וגרפים לפול
    plot_fold_results(fold + 1, history_current_fold, fold_val_labels, fold_val_preds, 
                      fold_new_data_labels, fold_new_data_preds, current_fold_new_data_recall)

    fold_results_effnet.append({'fold': fold+1, 'best_val_acc': best_val_acc, 'history': history_current_fold, 'new_data_recall': current_fold_new_data_recall if current_fold_new_data_recall != -1.0 else None})
    end_fold_time = time.time(); #print(f"Fold {fold+1} finished in {end_fold_time - start_fold_time:.2f} seconds.")


# --- סיכום תוצאות K-Fold ---
print(f"\n===== K-Fold Cross-Validation Summary ({MODEL_NAME} Frozen + Dropout) =====")
# ... (הדפסת סיכום ממוצעים ושמירת המודל הטוב ביותר לקובץ כמו קודם) ...
# ... (הצגת גרפים של הפול הטוב ביותר כמו קודם) ...

all_val_accs_eff = [r['best_val_acc'] for r in fold_results_effnet]
mean_val_acc_eff = np.mean(all_val_accs_eff) if all_val_accs_eff else 0; std_val_acc_eff = np.std(all_val_accs_eff) if all_val_accs_eff else 0
print(f"Average Validation Accuracy across {N_SPLITS} folds: {mean_val_acc_eff:.4f} +/- {std_val_acc_eff:.4f}")
recalls_eff = [r['new_data_recall'] for r in fold_results_effnet if r.get('new_data_recall') is not None and r['new_data_recall'] != -1.0]
if recalls_eff:
     mean_new_recall_eff = np.mean(recalls_eff); std_new_recall_eff = np.std(recalls_eff)
     print(f"Average 'Steps' Recall on New Dataset across {len(recalls_eff)} folds: {mean_new_recall_eff:.4f} +/- {std_new_recall_eff:.4f}")
else: print("No valid results for the new dataset recorded.")

if best_fold_model_state:
    print(f"\nBest overall model was found in Fold {best_fold_number}, Recall: {best_overall_new_data_recall:.4f}")
    best_model_filename = f"best_{MODEL_NAME}_frozen_dropout_model_fold{best_fold_number}.pth"
    # ... (קוד שמירת המודל) ...
    torch.save({ 'model_state_dict': best_fold_model_state, 'model_name': MODEL_NAME, 'dropout_p': DROPOUT_P,
                 'img_height': IMG_HEIGHT_FINAL, 'img_width': IMG_WIDTH_FINAL, 'n_mels': N_MELS, 'n_fft': N_FFT, 'hop_length': HOP_LENGTH,
                 'imagenet_mean': imagenet_mean, 'imagenet_std': imagenet_std }, best_model_filename)
    print(f"Best model state saved to: {best_model_filename}")
    if best_fold_history: # הצגת היסטוריית הפול הטוב ביותר
       plot_fold_results(best_fold_number, best_fold_history, [], [], None, None, best_overall_new_data_recall) 
else: print("\nNo best overall model was saved.")
```

```python
# ==============================================================================
# === קוד מלא ומתוקן סופית: חיפוש היפרפרמטרים עם Optuna, MobileNetV3, STFT ===
# ==============================================================================

print("--- Importing Libraries ---")
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
# <<< תיקון: ייבוא גם של Large Weights וגם Small Weights >>>
from torchvision.models import mobilenet_v3_small, MobileNet_V3_Small_Weights, mobilenet_v3_large, MobileNet_V3_Large_Weights
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import precision_score, recall_score, confusion_matrix, accuracy_score, classification_report
from scipy.signal import butter, filtfilt, spectrogram
import librosa
import librosa.display
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
import time
import random
from copy import deepcopy
import cv2         # <<< הוספה: ייבוא cv2 לעיבוד תמונה >>>
import optuna
import joblib      # <<< הוספה: ייבוא joblib לשמירת study >>>
import warnings
import heapq       # <<< הוספה: ייבוא heapq לבחירת top N >>>

warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

# --- 1. הגדרות גלובליות ---
print("--- Global Settings ---")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
LABEL_MAN = 0
LABEL_CAR = 1
LABEL_NOISE = 2
CLASS_NAMES = ["Steps", "Vehicle", "Noise"]
NUM_CLASSES = 3

# פרמטרים של עיבוד אות
window_size = 2000  # גודל חלון בדגימות
stride = 500      # גודל צעד בדגימות
fs = 1000.0       # קצב דגימה (Hz)
max_channels = 39 # מספר ערוצים מקסימלי (לריפוד)
lowcut = 1.0      # תדר חיתוך נמוך לפילטר (Hz)
highcut = 100.0   # תדר חיתוך גבוה לפילטר (Hz)
order = 4         # סדר הפילטר

# פרמטרים של STFT
N_FFT_STFT = 512
HOP_LENGTH_STFT = N_FFT_STFT // 4 # חפיפה של 75%
N_FREQ_BINS = N_FFT_STFT // 2 + 1 # מספר תדרים חיוביים

# פרמטרים של תמונה
IMG_HEIGHT_FINAL = 224 # גובה תמונת הספקטרוגרמה הסופית
IMG_WIDTH_FINAL = 224  # רוחב תמונת הספקטרוגרמה הסופית

# פרמטרים של Optuna ואימון
N_SPLITS_OPT = 5        # מספר folds ב-Cross Validation
EPOCHS_OPT = 25         # מספר מקסימלי של epochs לאימון כל fold
PATIENCE_OPT = 7        # מספר epochs להפסקת אימון מוקדמת אם אין שיפור
BATCH_SIZE_OPT = 32     # גודל batch
MODEL_NAME_OPT = 'mobilenet_v3_small' # או 'mobilenet_v3_large'
TIMEOUT_SECONDS_OPT = 4 * 3600 # זמן מקסימלי לחיפוש (4 שעות)
STUDY_NAME_OPT = f"geophone-{MODEL_NAME_OPT}-stft-search-v4" # עדכון גרסה בשם
STUDY_FILENAME = f"{STUDY_NAME_OPT}.pkl"
TOP_N_MODELS = 5        # מספר המודלים (trials) הטובים ביותר לשמור בסוף
MIN_VAL_ACC_THRESHOLD = 0.70  # סף מינימלי של דיוק ולידציה לשמירת מודל fold
MAX_VAL_ACC_THRESHOLD = 0.995 # סף מקסימלי (למניעת overfitting קיצוני או בעיות)

# נתיבי נתונים ושמירה
ORIGINAL_DATA_PATHS = {
    LABEL_MAN: {'path': 'man.csv'},
    LABEL_CAR: {'path': 'car.csv', 'args': {'skiprows': 1}}, # הנחה שהכותרת בשורה הראשונה
    LABEL_NOISE: {'path': 'nothing.csv'}
}
NEW_DATA_FOLDER = 'normalized_csv' # תיקייה עם קבצי CSV חדשים לבדיקה נוספת
SAVE_PATH_BASE = f"./saved_models_{MODEL_NAME_OPT}_stft_v4/" # עדכון גרסה בנתיב

# <<< תיקון: יצירת תיקיית שמירה אם לא קיימת >>>
os.makedirs(SAVE_PATH_BASE, exist_ok=True)

print(f"Using device: {device}")
print(f"Target Image Size: ({IMG_HEIGHT_FINAL}, {IMG_WIDTH_FINAL})")
print(f"Saving models to: {SAVE_PATH_BASE}")
print(f"Optuna study name: {STUDY_NAME_OPT}")

# --- 2. פונקציות עזר ועיבוד נתונים ---
print("\n--- Defining Helper Functions ---")

def pad_channels(data, target_channels):
    """מרפד או חותך את מספר הערוצים למספר היעד."""
    if data.ndim == 1:
        data = data.reshape(-1, 1) # הופך וקטור עמודה למטריצה עם עמודה אחת
    current_channels = data.shape[1]
    if current_channels == target_channels:
        return data
    elif current_channels < target_channels:
        padding = np.zeros((data.shape[0], target_channels - current_channels), dtype=data.dtype)
        return np.hstack((data, padding))
    else: # current_channels > target_channels
        return data[:, :target_channels]

def create_windows(signal, label, window_size=window_size, stride=stride):
    """יוצר חלונות מהאות עם תווית מתאימה."""
    X_windows, y_labels = [], []
    if signal.shape[0] < window_size:
        # print(f"Warning: Signal length ({signal.shape[0]}) is less than window size ({window_size}). Skipping.")
        return [], [] # <<< תיקון: החזר רשימות ריקות אם האות קצר מדי >>>
    for start in range(0, signal.shape[0] - window_size + 1, stride):
        window = signal[start : start + window_size, :]
        X_windows.append(window)
        y_labels.append(label)
    return X_windows, y_labels

def add_gaussian_noise_signal(signal_1d, noise_level=0.01):
    """מוסיף רעש גאוסיאני לאות חד-ממדי."""
    noise = np.random.normal(0, noise_level, signal_1d.shape)
    # <<< תיקון: המרה ל-float32 >>>
    return (signal_1d + noise).astype(np.float32)

def time_stretch_signal(signal_1d, stretch_rate=1.0, target_len=window_size):
    """מבצע מתיחת זמן על אות חד-ממדי ומבטיח אורך יעד."""
    try:
        # <<< תיקון: המרה ל-float32 לפני העברה ל-librosa >>>
        stretched = librosa.effects.time_stretch(signal_1d.astype(np.float32), rate=stretch_rate)
        current_len = len(stretched)

        if current_len == target_len:
            return stretched
        elif current_len > target_len:
            # חיתוך אקראי אם האות ארוך מדי
            start = random.randint(0, current_len - target_len)
            return stretched[start : start + target_len]
        else: # current_len < target_len
            # ריפוד אקראי אם האות קצר מדי
            pad_len = target_len - current_len
            pad_start = random.randint(0, pad_len)
            padded = np.zeros(target_len, dtype=np.float32)
            padded[pad_start : pad_start + current_len] = stretched
            return padded
    except Exception as e:
        # print(f"Warning: Time stretch failed with rate {stretch_rate}. Returning original. Error: {e}")
        # במקרה של שגיאה, החזר את האות המקורי באורך המטרה (עם ריפוד/חיתוך אם צריך)
        if len(signal_1d) == target_len:
            return signal_1d.astype(np.float32)
        elif len(signal_1d) > target_len:
            return signal_1d[:target_len].astype(np.float32)
        else:
            padded = np.zeros(target_len, dtype=np.float32)
            padded[:len(signal_1d)] = signal_1d
            return padded


def create_stft_spectrogram(window_1d, sr=int(fs), n_fft=N_FFT_STFT, hop_length=HOP_LENGTH_STFT):
    """
    יוצר ספקטרוגרמת STFT מאות חד-ממדי בחלון, כולל סינון תדרים ונרמול.
    מחזיר ספקטרוגרמה מנורמלת או None במקרה של שגיאה/קלט לא תקין.
    """
    try:
        # <<< תיקון: ודא שהקלט הוא float32 >>>
        signal = window_1d.astype(np.float32)

        # בדוק אם האות מכיל רק אפסים או ערכים קבועים (שעלולים לגרום לשגיאות)
        if np.all(signal == signal[0]):
           # print("Warning: Constant signal detected in window. Returning zero spectrogram.")
           return np.zeros((min(N_FREQ_BINS // 2, 128), 10), dtype=np.float32) # החזר מערך קטן של אפסים

        # חשב STFT באמצעות scipy.signal.spectrogram
        # nperseg = גודל חלון STFT, noverlap = חפיפה, nfft = מספר נקודות FFT
        f, t, Sxx = spectrogram(signal, fs=sr, nperseg=n_fft, noverlap=n_fft - hop_length,
                                nfft=n_fft, detrend=False, scaling='spectrum')

        # סנן תדרים מעל highcut (עם מרווח קטן)
        freq_slice = np.where(f <= highcut * 1.1)[0]
        if len(freq_slice) == 0:
            # print(f"Warning: No frequency bins found below {highcut * 1.1} Hz. Returning zero spec.")
            return np.zeros((min(N_FREQ_BINS // 2, 128), len(t) if len(t)>0 else 10), dtype=np.float32)

        f = f[freq_slice]
        Sxx = Sxx[freq_slice, :]

        # בדוק אם הספקטרוגרמה ריקה אחרי החיתוך
        if Sxx.size == 0:
            # print("Warning: Spectrogram is empty after frequency slicing. Returning zero spec.")
            return np.zeros((len(f) if len(f)>0 else min(N_FREQ_BINS // 2, 128), 10), dtype=np.float32)

        # המרה ל-dB ונרמול
        Sxx_db = 10 * np.log10(Sxx + 1e-9)  # הוספת epsilon קטן למניעת log(0)

        # נרמול בין 0 ל-1 באמצעות אחוזונים (יותר עמיד לערכים חריגים)
        min_db, max_db = np.percentile(Sxx_db, 5), np.percentile(Sxx_db, 95)
        if max_db <= min_db: # טפל במקרה שהערכים כמעט זהים
            max_db = min_db + 1e-6

        spec_norm = np.clip((Sxx_db - min_db) / (max_db - min_db), 0, 1)

        # <<< תיקון: ודא שהפלט הוא float32 >>>
        return spec_norm.astype(np.float32)

    except Exception as e:
        # print(f"Error creating STFT spectrogram: {e}. Returning zero spec.")
        # <<< תיקון: החזר מערך אפסים קטן ומוגדר במקרה של שגיאה >>>
        return np.zeros((min(N_FREQ_BINS // 2, 128), 10), dtype=np.float32)


def load_data_stft_specs(file_map, apply_signal_aug=False, signal_aug_prob=0.5, target_img_size=(IMG_HEIGHT_FINAL, IMG_WIDTH_FINAL)):
    """
    טוען נתונים מקבצי CSV, מייצר חלונות, מבצע אופציונלית אוגמנטציה לאות,
    יוצר ספקטרוגרמות STFT, משנה גודל, ומרכיב לרשימות numpy.
    """
    all_specs, all_labels = [], []
    try:
        # יצירת הפילטר פעם אחת מחוץ ללולאה
        b, a = butter(order, [lowcut/(fs/2), highcut/(fs/2)], btype='band')
    except ValueError as e:
        print(f"CRITICAL ERROR: Failed to create Butterworth filter: {e}. Check lowcut/highcut/fs values. Aborting load.")
        return np.array([]), np.array([]) # החזר מערכים ריקים

    print(f"Loading data & creating STFT (Signal Aug: {apply_signal_aug}, Prob: {signal_aug_prob})...")
    file_count = 0
    specs_created_count = 0
    processed_files = 0

    for label, info in file_map.items():
        filepaths = info.get('path', []) # קבל נתיב או רשימה ריקה
        extra_args = info.get('args', {})

        # ודא ש-filepaths הוא רשימה
        if isinstance(filepaths, str):
            filepaths = [filepaths]

        if not filepaths: # דלג אם אין קבצים עבור התווית הזו
            continue

        file_count += len(filepaths)

        for filepath in filepaths:
            if not os.path.exists(filepath):
                print(f"Warning: File not found: {filepath}. Skipping.")
                continue

            try:
                # נסה לקרוא את ה-CSV, עם טיפול בשגיאות נפוצות (כותרת, רווחים)
                try:
                    df = pd.read_csv(filepath, header=None, skipinitialspace=True, **extra_args)
                except pd.errors.EmptyDataError:
                    print(f"Warning: Empty file: {filepath}. Skipping.")
                    continue
                except Exception as e_read:
                    # נסה שוב עם דילוג על שורה ראשונה אם הקריאה הראשונה נכשלה
                    print(f"Info: Reading {filepath} failed, trying with skiprows=1. Error: {e_read}")
                    try:
                        df = pd.read_csv(filepath, header=None, skiprows=1, skipinitialspace=True, **extra_args)
                    except pd.errors.EmptyDataError:
                        print(f"Warning: Empty file even with skiprows=1: {filepath}. Skipping.")
                        continue
                    except Exception as e_read2:
                        print(f"CRITICAL WARNING: Failed to read CSV file: {filepath}. Error: {e_read2}. Skipping.")
                        continue

                df = df.dropna() # הסר שורות עם ערכים חסרים
                if df.empty:
                    # print(f"Warning: File became empty after dropna: {filepath}. Skipping.")
                    continue

                data = df.values
                # ודא שהנתונים מספריים
                if not np.issubdtype(data.dtype, np.number):
                    try:
                        data = data.astype(float)
                    except ValueError:
                        print(f"Warning: Could not convert data to numeric in file: {filepath}. Skipping.")
                        continue

                # בדוק אם יש מספיק נתונים ומימדים תקינים
                if data.shape[0] < window_size or data.ndim < 2 or data.shape[1] == 0:
                    # print(f"Warning: Insufficient data or invalid shape in {filepath} after processing. Shape: {data.shape}. Skipping.")
                    continue

                # רפד ערוצים אם יש פחות מ-max_channels
                data_padded = pad_channels(data, max_channels)

                # יצירת חלונות
                windows, labels_w = create_windows(data_padded, label)
                if not windows: # אם לא נוצרו חלונות (למשל, אות קצר מדי)
                    continue

                # עיבוד כל חלון
                for i, w in enumerate(windows):
                    try:
                        # החל פילטר Bandpass (על כל הערוצים, אבל נשתמש רק בראשון ל-STFT)
                        # שימוש ב-filtfilt למניעת פאזה
                        filtered_w = filtfilt(b, a, w, axis=0)

                        # השתמש רק בערוץ הראשון ליצירת הספקטרוגרמה
                        signal_to_process = filtered_w[:, 0]

                        # אוגמנטציית אות (רעש או מתיחה) - לפני STFT
                        if apply_signal_aug and random.random() < signal_aug_prob:
                            aug_type = random.choice(['stretch', 'noise'])
                            if aug_type == 'stretch':
                                # מתיחת זמן בשיעור אקראי קטן
                                signal_to_process = time_stretch_signal(signal_to_process, stretch_rate=np.random.uniform(0.9, 1.1), target_len=window_size)
                            elif aug_type == 'noise':
                                # הוספת רעש גאוסיאני ברמה אקראית קטנה
                                signal_to_process = add_gaussian_noise_signal(signal_to_process, noise_level=np.random.uniform(0.005, 0.015))

                        # יצירת ספקטרוגרמת STFT מנורמלת
                        spec_norm = create_stft_spectrogram(signal_to_process)

                        # <<< תיקון: בדיקה קפדנית של הפלט מ-create_stft_spectrogram >>>
                        if spec_norm is None or spec_norm.shape[0] < 5 or spec_norm.shape[1] < 5: # בדוק אם הספקטרוגרמה תקינה ובעלת גודל מינימלי
                            # print(f"Warning: Invalid or too small spectrogram generated for a window from {filepath}. Skipping window.")
                            continue

                        # שינוי גודל הספקטרוגרמה לגודל היעד של המודל
                        try:
                            resized_spec = cv2.resize(spec_norm, (target_img_size[1], target_img_size[0]), interpolation=cv2.INTER_LINEAR)
                        except cv2.error as e_resize:
                            print(f"Warning: cv2.resize failed for a window from {filepath}. Spec shape: {spec_norm.shape}. Error: {e_resize}. Skipping window.")
                            continue

                        # שכפול ל-3 ערוצים (כמו תמונת RGB) עבור המודל
                        spec_3_channels = np.stack([resized_spec] * 3, axis=0) # צורת הפלט: (3, H, W)

                        # <<< תיקון: בדיקה סופית לערכים אינסופיים או NaN >>>
                        if not np.isfinite(spec_3_channels).all():
                            # print(f"Warning: Non-finite values (inf/NaN) found in final spectrogram for a window from {filepath}. Skipping window.")
                            continue

                        all_specs.append(spec_3_channels)
                        all_labels.append(labels_w[i])
                        specs_created_count += 1

                    except Exception as e_window:
                        # תפיסת שגיאות ספציפיות לחלון (למשל, ב-filtfilt, augment, stft)
                        print(f"Warning: Error processing a window from {filepath}: {e_window}. Skipping window.")
                        pass # המשך לחלון הבא

            except Exception as e_file:
                # תפיסת שגיאות ברמת הקובץ (למשל, קריאה, עיבוד ראשוני)
                print(f"Warning: Error processing file {filepath}: {e_file}. Skipping file.")
                pass # המשך לקובץ הבא
            finally:
                processed_files += 1
                if processed_files % 50 == 0:
                    print(f"  Processed {processed_files}/{file_count} files...")


    print(f"Finished loading. Processed {processed_files} files. Total STFT spectrograms created: {specs_created_count}")

    # <<< תיקון: החזר מערכי numpy גם אם הם ריקים >>>
    if not all_specs:
        print("Warning: No spectrograms were successfully created.")
        return np.array([]), np.array([], dtype=np.int64)

    # המר רשימות למערכי numpy בסוף התהליך
    return np.array(all_specs, dtype=np.float32), np.array(all_labels, dtype=np.int64)

# --- טעינת הנתונים ---
print("\n--- Loading All Data as STFT Spectrograms ---")

# טעינת נתונים מקוריים (ללא אוגמנטציה בשלב זה)
all_original_specs_np, all_original_labels_np = load_data_stft_specs(ORIGINAL_DATA_PATHS, apply_signal_aug=False)

# טעינת נתונים חדשים (אם קיימים)
# <<< תיקון: בדיקת קיום קבצים ויצירת מילון נכון >>>
new_specs_np = np.array([])
new_labels_np = np.array([], dtype=np.int64)
if os.path.exists(NEW_DATA_FOLDER):
    new_files = glob.glob(os.path.join(NEW_DATA_FOLDER, '*.csv'))
    if new_files:
        print(f"Found {len(new_files)} new CSV files in '{NEW_DATA_FOLDER}'.")
        # הנחה שכל הקבצים החדשים הם מסוג LABEL_MAN (צעדים) - שנה לפי הצורך
        new_file_map = {LABEL_MAN: {'path': new_files}}
        new_specs_np, new_labels_np = load_data_stft_specs(new_file_map, apply_signal_aug=False)
    else:
        print(f"No CSV files found in '{NEW_DATA_FOLDER}'.")
else:
    print(f"Directory '{NEW_DATA_FOLDER}' not found. Skipping loading new data.")


# בדיקה קריטית אם הנתונים המקוריים נטענו
if all_original_specs_np.size == 0:
    print("\n!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
    print("CRITICAL ERROR: Failed to load any original data. Cannot proceed.")
    print("Please check file paths, file formats, and data processing steps.")
    print("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!\n")
    exit() # יציאה מהסקריפט

print("Original STFT Spectrogram data shape:", all_original_specs_np.shape, "Labels shape:", all_original_labels_np.shape)
if new_specs_np.size > 0:
    print("New STFT Spectrogram data shape:", new_specs_np.shape, "Labels shape:", new_labels_np.shape)

# הצגת התפלגות התוויות
unique_orig, counts_orig = np.unique(all_original_labels_np, return_counts=True)
print("Original data label distribution:", dict(zip(unique_orig, counts_orig)))
if new_labels_np.size > 0:
    unique_new, counts_new = np.unique(new_labels_np, return_counts=True)
    print("New data label distribution:", dict(zip(unique_new, counts_new)))


# --- 3. הגדרת טרנספורמציות, Dataset, מודל, פונקציות אימון/הערכה ---
print("\n--- Defining Transforms, Dataset, Model, Train/Eval Functions, Plotting ---")

# סטטיסטיקות נרמול של ImageNet (מתאימות למודלים שאומנו מראש)
imagenet_mean = [0.485, 0.456, 0.406]
imagenet_std = [0.229, 0.224, 0.225]
print("ImageNet normalization stats defined.")

# אוגמנטציית SpecAugment (מבוצעת על הטנזור ב-GPU)
class SpecAugment(nn.Module):
    def __init__(self, freq_mask_param=20, time_mask_param=40, num_freq_masks=1, num_time_masks=1):
        super().__init__()
        self.freq_mask_param = freq_mask_param
        self.time_mask_param = time_mask_param
        self.num_freq_masks = num_freq_masks
        self.num_time_masks = num_time_masks

    def forward(self, spec):
        # <<< תיקון: בדיקת סוג הקלט >>>
        if not isinstance(spec, torch.Tensor):
            raise TypeError(f"Input must be a Tensor, got {type(spec)}")
        if spec.ndim != 3:
             raise ValueError(f"Input must be a 3D tensor (C, F, T), got {spec.ndim} dimensions")

        spec_aug = spec.clone()
        _, F, T = spec_aug.shape # C, Freq, Time

        # <<< תיקון: ודא שפרמטרי המסכה אינם גדולים ממימדי הספקטרוגרמה >>>
        fm_param = min(self.freq_mask_param, F // 2) # הגבל מקסימום לחצי מגובה התדר
        tm_param = min(self.time_mask_param, T // 2) # הגבל מקסימום לחצי מרוחב הזמן

        # החל מסכות תדר
        for _ in range(self.num_freq_masks):
            if fm_param > 0: # רק אם הפרמטר חיובי
                f = random.randrange(0, fm_param) # רוחב המסכה
                if F - f > 0: # רק אם רוחב המסכה קטן מהגובה
                   f0 = random.randrange(0, F - f) # מיקום התחלה
                   spec_aug[:, f0:f0+f, :] = spec.mean() # אפס את האזור (או השתמש בממוצע)

        # החל מסכות זמן
        for _ in range(self.num_time_masks):
            if tm_param > 0: # רק אם הפרמטר חיובי
                t = random.randrange(0, tm_param) # רוחב המסכה
                if T - t > 0: # רק אם רוחב המסכה קטן מהרוחב
                    t0 = random.randrange(0, T - t) # מיקום התחלה
                    spec_aug[:, :, t0:t0+t] = spec.mean() # אפס את האזור

        return spec_aug

# הרכבת הטרנספורמציות
train_transforms = transforms.Compose([
    # SpecAugment מוחל ראשון (על הטנזור לפני הנרמול)
    SpecAugment(),
    # נרמול לפי ImageNet
    transforms.Normalize(mean=imagenet_mean, std=imagenet_std)
])

val_test_transforms = transforms.Compose([
    # רק נרמול עבור ולידציה ובדיקה
    transforms.Normalize(mean=imagenet_mean, std=imagenet_std)
])
print("Transforms defined (Train includes SpecAugment).")

# הגדרת Dataset של PyTorch
class SpectrogramDataset(Dataset):
    def __init__(self, specs_np, labels_np, transform=None):
        # <<< תיקון: המרה לטנזורים ב-init לחסכון בזיכרון וזמן >>>
        if not isinstance(specs_np, np.ndarray) or not isinstance(labels_np, np.ndarray):
             raise TypeError("Input specs and labels must be numpy arrays.")
        # ודא שהסוגים נכונים לפני ההמרה
        self.specs = torch.from_numpy(specs_np.astype(np.float32))
        self.labels = torch.from_numpy(labels_np.astype(np.int64))
        self.transform = transform
        if len(self.specs) != len(self.labels):
             raise ValueError("Number of specs and labels must match.")

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        spec = self.specs[idx]
        label = self.labels[idx]
        # החלת טרנספורמציות (כמו SpecAugment ונרמול) אם קיימות
        if self.transform:
            spec = self.transform(spec)
        return spec, label
print("Dataset defined.")

# פונקציה להגדרת מודל MobileNetV3
# <<< תיקון: הוסר פרמטר dropout_p שאינו בשימוש בקונסטרוקטור >>>
def setup_mobilenet_model(model_name='mobilenet_v3_small', pretrained=True, num_classes=NUM_CLASSES):
    """מגדיר מודל MobileNetV3 (קטן או גדול) עם אופציה לאימון מראש."""
    if model_name == 'mobilenet_v3_small':
        weights = MobileNet_V3_Small_Weights.DEFAULT if pretrained else None
        model = mobilenet_v3_small(weights=weights)
        # החלפת השכבה הלינארית האחרונה ב-classifier
        num_ftrs = model.classifier[3].in_features
        # החלף רק את השכבה האחרונה, שמור על השכבות הקודמות ב-classifier אם תרצה
        model.classifier[3] = nn.Linear(num_ftrs, num_classes)
        print(f"Using {model_name} with {num_ftrs} input features to final layer.")
    elif model_name == 'mobilenet_v3_large':
        weights = MobileNet_V3_Large_Weights.DEFAULT if pretrained else None
        model = mobilenet_v3_large(weights=weights)
        # החלפת השכבה הלינארית האחרונה ב-classifier
        num_ftrs = model.classifier[3].in_features
        model.classifier[3] = nn.Linear(num_ftrs, num_classes)
        print(f"Using {model_name} with {num_ftrs} input features to final layer.")
    else:
        raise ValueError(f"Unsupported model name: {model_name}. Choose 'mobilenet_v3_small' or 'mobilenet_v3_large'.")

    # הקפאת שכבות ה-features אם משתמשים במשקלים מאומנים מראש
    if pretrained:
        print("Freezing feature extractor layers.")
        for param in model.features.parameters():
            param.requires_grad = False
        # ודא ששכבות ה-classifier כן ניתנות לאימון
        for param in model.classifier.parameters():
             param.requires_grad = True
    else:
        print("Training all layers from scratch (or fine-tuning if weights were loaded manually).")

    # העבר מודל ל-GPU אם זמין
    model = model.to(device)
    return model
print("Model setup function (MobileNetV3) defined.")

# פונקציית אימון ל-epoch אחד
def train_epoch_model(model, dataloader, criterion, optimizer, device):
    model.train()  # העבר למצב אימון
    running_loss = 0.0
    correct_preds = 0
    total_samples = 0
    batch_num = 0

    for inputs, labels in dataloader:
        batch_num += 1
        outputs = None # <<< אתחול לטיפול בשגיאות >>>
        preds = None   # <<< אתחול לטיפול בשגיאות >>>
        try:
            inputs, labels = inputs.to(device), labels.to(device)

            # אפס גרדיאנטים
            optimizer.zero_grad()

            # קדימה (forward pass)
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            # <<< תיקון: בדוק אם ה-loss הוא NaN לפני backward >>>
            if loss is None or torch.isnan(loss):
                print(f"  Warning: NaN loss detected during training batch {batch_num}. Skipping batch.")
                continue # דלג על ה-batch הזה

            # חישוב ניבויים
            preds = outputs.max(1)[1]

            # אחורה (backward pass) ועדכון משקלים
            loss.backward()
            optimizer.step()

            # סטטיסטיקה
            running_loss += loss.item() * inputs.size(0)
            total_samples += labels.size(0)
            # <<< תיקון: ודא ש-preds לא None לפני חישוב correct_preds >>>
            if preds is not None:
                correct_preds += (preds == labels).sum().item()

        # <<< תיקון: תפיסת שגיאות ברמת ה-batch >>>
        except Exception as e:
            print(f"  ERROR during training batch {batch_num}: {e}")
            # נסה להמשיך ל-batch הבא
            continue

    # <<< תיקון: טפל במקרה ש-total_samples הוא 0 >>>
    if total_samples == 0:
        print("  Warning: No samples processed in train_epoch_model.")
        return 0.0, 0.0

    epoch_loss = running_loss / total_samples
    epoch_acc = correct_preds / total_samples
    return epoch_loss, epoch_acc

# פונקציית הערכה (evaluation)
def evaluate_model(model, dataloader, criterion, device):
    model.eval()   # העבר למצב הערכה
    running_loss = 0.0
    correct_preds = 0
    total_samples = 0
    all_preds = []   # <<< הוספה: לאיסוף כל הניבויים >>>
    all_labels = []  # <<< הוספה: לאיסוף כל התוויות האמיתיות >>>

    with torch.no_grad(): # אין צורך לחשב גרדיאנטים בהערכה
        batch_num = 0
        for inputs, labels in dataloader:
            batch_num += 1
            outputs = None # <<< אתחול לטיפול בשגיאות >>>
            preds = None   # <<< אתחול לטיפול בשגיאות >>>
            try:
                inputs, labels = inputs.to(device), labels.to(device)

                # קדימה
                outputs = model(inputs)
                loss = criterion(outputs, labels)

                # <<< תיקון: בדוק אם ה-loss הוא NaN >>>
                if loss is None or torch.isnan(loss):
                    print(f"  Warning: NaN loss detected during evaluation batch {batch_num}. Skipping batch.")
                    continue

                # חישוב ניבויים
                preds = outputs.max(1)[1]

                # סטטיסטיקה
                running_loss += loss.item() * inputs.size(0)
                total_samples += labels.size(0)

                # <<< תיקון: אסוף תוויות וניבויים >>>
                if preds is not None:
                    all_preds.extend(preds.cpu().numpy())
                    all_labels.extend(labels.cpu().numpy())
                    correct_preds += (preds == labels).sum().item()

            # <<< תיקון: תפיסת שגיאות ברמת ה-batch >>>
            except Exception as e:
                print(f"  ERROR during evaluation batch {batch_num}: {e}")
                # נסה להמשיך ל-batch הבא
                continue

    # <<< תיקון: טפל במקרה ש-total_samples הוא 0 >>>
    if total_samples == 0:
        print("  Warning: No samples processed in evaluate_model.")
        return 0.0, 0.0, [], [] # החזר רשימות ריקות

    epoch_loss = running_loss / total_samples
    epoch_acc = correct_preds / total_samples
    # <<< תיקון: החזר גם את התוויות והניבויים >>>
    return epoch_loss, epoch_acc, all_labels, all_preds
print("Train/Eval functions defined.")

# --- פונקציית הצגת גרפים (נשארת זהה, אך תוכן תלוי בנתונים שנשמרים) ---
def plot_fold_results(fold_num, history, val_labels, val_preds,
                      new_data_labels=None, new_data_preds=None, new_data_recall=None,
                      class_names=CLASS_NAMES, save_path=None, trial_num=None):
    """מציגה גרפים של אימון, מטריצת בלבול ודוח סיווג."""
    if not history or not val_labels or not val_preds:
        print(f"  Plotting skipped for Fold {fold_num} due to missing data.")
        return

    try:
        fig, axes = plt.subplots(1, 3, figsize=(24, 6))
        prefix = f"Trial_{trial_num}_Fold_{fold_num}" if trial_num is not None else f"Fold_{fold_num}"

        # גרף Loss
        axes[0].plot(history.get('train_loss', []), label='Train Loss')
        axes[0].plot(history.get('val_loss', []), label='Validation Loss')
        axes[0].set_title(f'{prefix} - Loss vs. Epochs')
        axes[0].set_xlabel('Epochs')
        axes[0].set_ylabel('Loss')
        axes[0].legend()
        axes[0].grid(True)

        # גרף Accuracy
        axes[1].plot(history.get('train_acc', []), label='Train Accuracy')
        axes[1].plot(history.get('val_acc', []), label='Validation Accuracy')
        axes[1].set_title(f'{prefix} - Accuracy vs. Epochs')
        axes[1].set_xlabel('Epochs')
        axes[1].set_ylabel('Accuracy')
        axes[1].legend()
        axes[1].grid(True)

        # מטריצת בלבול (Validation Data)
        cm = confusion_matrix(val_labels, val_preds)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names, ax=axes[2])
        axes[2].set_title(f'{prefix} - Validation Confusion Matrix')
        axes[2].set_xlabel('Predicted Label')
        axes[2].set_ylabel('True Label')

        plt.tight_layout()
        if save_path:
            plot_filename = os.path.join(save_path, f"{prefix}_training_plots.png")
            plt.savefig(plot_filename)
            print(f"  Saved training plots to {plot_filename}")
        # plt.show() # אפשר להסיר אם רצים הרבה ניסויים
        plt.close(fig) # סגור את התמונה כדי לשחרר זיכרון

        # הדפסת דוח סיווג (Validation Data)
        print(f"\n--- {prefix} - Validation Classification Report ---")
        # בדוק אם יש מספיק תוויות/ניבויים ושהם לא קבועים
        if len(np.unique(val_labels)) > 1 and len(np.unique(val_preds)) > 1:
            print(classification_report(val_labels, val_preds, target_names=class_names, zero_division=0))
        else:
            print("  Could not generate classification report (likely due to constant labels/predictions).")
            print(f"  Val Acc: {accuracy_score(val_labels, val_preds):.4f}")


        # הצגת תוצאות על הנתונים החדשים (אם קיימים)
        if new_data_labels is not None and new_data_preds is not None:
            print(f"\n--- {prefix} - New Data Classification Report ---")
            if len(np.unique(new_data_labels)) > 1 and len(np.unique(new_data_preds)) > 1:
               print(classification_report(new_data_labels, new_data_preds, target_names=class_names, zero_division=0))
            else:
               print("  Could not generate classification report for new data.")
               print(f"  New Data Acc: {accuracy_score(new_data_labels, new_data_preds):.4f}")

            if new_data_recall is not None:
                print(f"  Recall for '{class_names[LABEL_MAN]}' on New Data: {new_data_recall:.4f}")

            # גרף מטריצת בלבול נוסף עבור הנתונים החדשים
            try:
                fig_new, ax_new = plt.subplots(1, 1, figsize=(8, 6))
                cm_new = confusion_matrix(new_data_labels, new_data_preds)
                sns.heatmap(cm_new, annot=True, fmt='d', cmap='Greens', xticklabels=class_names, yticklabels=class_names, ax=ax_new)
                ax_new.set_title(f'{prefix} - New Data Confusion Matrix')
                ax_new.set_xlabel('Predicted Label')
                ax_new.set_ylabel('True Label')
                plt.tight_layout()
                if save_path:
                    plot_filename_new = os.path.join(save_path, f"{prefix}_new_data_confusion_matrix.png")
                    plt.savefig(plot_filename_new)
                    print(f"  Saved new data confusion matrix to {plot_filename_new}")
                # plt.show()
                plt.close(fig_new)
            except Exception as e_plot_new:
                print(f"  Warning: Could not plot new data confusion matrix: {e_plot_new}")

    except Exception as e:
        print(f"An error occurred during plotting for Fold {fold_num}: {e}")

print("Plotting function defined.")


# --- 4. פונקציית המטרה (Objective) ל-Optuna ---
print("\n--- Defining Optuna Objective Function ---")

# משתנה גלובלי לאחסון המודלים הטובים ביותר מכל ה-trials (אפשר גם להשתמש ב-user_attrs)
# זה לא הכרחי אם משתמשים ב-user_attrs בצורה נכונה
# top_n_trial_results = []

def objective(trial):
    """פונקציית המטרה ש-Optuna ימזער/ימקסם."""
    start_trial_time = time.time()
    print(f"\n{'='*15} Optuna Trial {trial.number} Starting {'='*15}")

    # --- 1. דגימת היפרפרמטרים ---
    lr = trial.suggest_float("lr", 1e-4, 1e-2, log=True)
    # <<< הערה: dropout_p נדגם אך לא מועבר ל-setup_mobilenet_model כרגע >>>
    dropout_p_ignored = trial.suggest_float("dropout_p", 0.2, 0.6, step=0.1)
    weight_decay = trial.suggest_float("weight_decay", 5e-5, 5e-3, log=True)
    # פרמטרים של SpecAugment
    specaug_freq = trial.suggest_int("specaug_freq", 10, 30)
    specaug_time = trial.suggest_int("specaug_time", 20, 60)
    specaug_f_masks = trial.suggest_int("specaug_f_masks", 1, 2)
    specaug_t_masks = trial.suggest_int("specaug_t_masks", 1, 2)

    print(f"  Trial Params: LR={lr:.1e}, WD={weight_decay:.1e}, "
          f"SpecAugF={specaug_freq}, SpecAugT={specaug_time}, "
          f"N_F={specaug_f_masks}, N_T={specaug_t_masks}")
          # f"Dropout(ignored)={dropout_p_ignored:.1f}") # הדפסה שהפרמטר לא בשימוש

    # --- 2. הגדרת טרנספורמציות אימון ספציפיות ל-trial ---
    current_train_transforms = transforms.Compose([
        SpecAugment(freq_mask_param=specaug_freq, time_mask_param=specaug_time,
                    num_freq_masks=specaug_f_masks, num_time_masks=specaug_t_masks),
        transforms.Normalize(mean=imagenet_mean, std=imagenet_std)
    ])

    # --- 3. Cross-Validation ---
    # <<< תיקון: random_state משתנה לכל trial למניעת חזרתיות >>>
    kfold = StratifiedKFold(n_splits=N_SPLITS_OPT, shuffle=True, random_state=trial.number + 42) # הוספת מספר ה-trial ל-seed
    fold_recalls_new_data = [] # לאחסון Recall על הנתונים החדשים מכל fold שעמד בתנאים
    fold_val_accs = []         # לאחסון Validation Accuracy מכל fold
    fold_best_models_info = [] # לאחסון מידע על המודלים הטובים ביותר מכל fold שעמד בתנאים

    # ודא שקיימים נתונים מקוריים
    if all_original_specs_np.size == 0:
        print("  ERROR: Original data is empty. Cannot perform cross-validation. Returning 0.")
        return 0.0 # החזר ערך נמוך כדי ש-Optuna ידלג

    for fold, (train_ids, val_ids) in enumerate(kfold.split(all_original_specs_np, all_original_labels_np)):
        fold_start_time = time.time()
        print(f"  --- Fold {fold+1}/{N_SPLITS_OPT} ---")

        # --- 3.1 הכנת נתונים ל-fold ---
        X_train_fold, y_train_fold = all_original_specs_np[train_ids], all_original_labels_np[train_ids]
        X_val_fold, y_val_fold = all_original_specs_np[val_ids], all_original_labels_np[val_ids]

        # יצירת Dataset ו-DataLoader
        train_dataset = SpectrogramDataset(X_train_fold, y_train_fold, transform=current_train_transforms)
        val_dataset = SpectrogramDataset(X_val_fold, y_val_fold, transform=val_test_transforms)

        # num_workers=0 עשוי להיות יציב יותר, במיוחד ב-Windows או עם שגיאות זיכרון
        train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE_OPT, shuffle=True, num_workers=0, pin_memory=True)
        val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE_OPT, shuffle=False, num_workers=0, pin_memory=True)

        # --- 3.2 הגדרת מודל, קריטריון ואופטימייזר ---
        # <<< תיקון: יצירת מודל חדש עבור כל fold >>>
        # <<< הערה: dropout_p_ignored אינו מועבר לפונקציה >>>
        model = setup_mobilenet_model(model_name=MODEL_NAME_OPT, pretrained=True, num_classes=NUM_CLASSES)
        criterion = nn.CrossEntropyLoss()

        # אופטימיזציה רק של הפרמטרים ב-classifier (כי ה-features מוקפאים)
        # אם pretrained=False, צריך להעביר את model.parameters()
        params_to_optimize = model.classifier.parameters() if any(p.requires_grad for p in model.features.parameters()) == False else model.parameters()
        optimizer = optim.AdamW(params_to_optimize, lr=lr, weight_decay=weight_decay)

        # --- 3.3 אימון ה-fold עם Early Stopping ---
        best_val_acc_fold = 0.0
        current_best_fold_model_state = None
        epochs_no_improve = 0
        best_epoch = -1
        history_current_fold_obj = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
        fold_val_labels_final = [] # לשמור את תוויות/ניבויי הולידציה של ה-epoch הטוב ביותר
        fold_val_preds_final = []

        for epoch in range(1, EPOCHS_OPT + 1):
            epoch_start_time = time.time()

            # אימון
            train_loss, train_acc = train_epoch_model(model, train_loader, criterion, optimizer, device)
            # הערכה
            val_loss, val_acc, fold_val_labels_epoch, fold_val_preds_epoch = evaluate_model(model, val_loader, criterion, device)

            print(f"    Epoch {epoch}/{EPOCHS_OPT} | Train Loss: {train_loss:.4f}, Acc: {train_acc:.4f} | "
                  f"Val Loss: {val_loss:.4f}, Acc: {val_acc:.4f} | Time: {time.time() - epoch_start_time:.1f}s")

            # שמירת היסטוריה
            history_current_fold_obj['train_loss'].append(train_loss)
            history_current_fold_obj['train_acc'].append(train_acc)
            history_current_fold_obj['val_loss'].append(val_loss)
            history_current_fold_obj['val_acc'].append(val_acc)

            # בדיקת Early Stopping
            if val_acc > best_val_acc_fold:
                best_val_acc_fold = val_acc
                # <<< תיקון: שימוש ב-deepcopy לשמירת מצב המודל ולא רק הפנייה >>>
                current_best_fold_model_state = deepcopy(model.state_dict())
                epochs_no_improve = 0
                best_epoch = epoch
                # שמור את התוויות והניבויים מה-epoch הטוב ביותר
                fold_val_labels_final = fold_val_labels_epoch
                fold_val_preds_final = fold_val_preds_epoch
                print(f"      -> New best validation accuracy: {best_val_acc_fold:.4f} at epoch {epoch}")
            else:
                epochs_no_improve += 1

            if epochs_no_improve >= PATIENCE_OPT:
                print(f"    Early stopping triggered at epoch {epoch}. Best val_acc: {best_val_acc_fold:.4f} from epoch {best_epoch}.")
                break # סיים את לולאת ה-epochs עבור ה-fold הזה

        # --- 3.4 הערכה על נתונים חדשים (אם רלוונטי) ---
        current_fold_new_data_recall = 0.0
        fold_new_data_labels = None
        fold_new_data_preds = None

        # הערך על הנתונים החדשים רק אם:
        # 1. נמצא מודל טוב ב-fold זה (current_best_fold_model_state אינו None)
        # 2. ה-validation accuracy של המודל הטוב עומד בספים שהוגדרו
        # 3. קיימים נתונים חדשים לטעינה (new_specs_np.size > 0)
        if current_best_fold_model_state and new_specs_np.size > 0:
            if MIN_VAL_ACC_THRESHOLD <= best_val_acc_fold <= MAX_VAL_ACC_THRESHOLD:
                print(f"    Fold {fold+1} best model (ValAcc={best_val_acc_fold:.4f}) qualifies for new data evaluation.")
                # טען את המצב הטוב ביותר למודל
                model.load_state_dict(current_best_fold_model_state)

                # הכן DataLoader לנתונים החדשים
                new_dataset = SpectrogramDataset(new_specs_np, new_labels_np, transform=val_test_transforms)
                new_loader = DataLoader(new_dataset, batch_size=BATCH_SIZE_OPT, shuffle=False, num_workers=0, pin_memory=True)

                # הערך את המודל על הנתונים החדשים
                try:
                    _, new_acc, fold_new_data_labels, fold_new_data_preds = evaluate_model(model, new_loader, criterion, device)
                    # חשב Recall ספציפית למחלקת 'צעדים' (LABEL_MAN)
                    current_fold_new_data_recall = recall_score(fold_new_data_labels, fold_new_data_preds,
                                                              labels=[LABEL_MAN], average='micro', zero_division=0) # או 'binary' אם זו המחלקה היחידה
                    print(f"      -> New Data Eval: Accuracy={new_acc:.4f}, Recall(Steps)={current_fold_new_data_recall:.4f}")

                    # שמור את פרטי ה-fold המוצלח הזה
                    fold_best_models_info.append({
                        'fold': fold + 1,
                        'val_acc': best_val_acc_fold,
                        'new_recall': current_fold_new_data_recall,
                        'epoch': best_epoch,
                        'state_dict': current_best_fold_model_state, # שמור את המצב עצמו
                        'history': history_current_fold_obj,
                        'val_labels': fold_val_labels_final,
                        'val_preds': fold_val_preds_final,
                        'new_labels': fold_new_data_labels,
                        'new_preds': fold_new_data_preds,
                    })

                except Exception as e_eval_new:
                    print(f"      ERROR during new data evaluation for fold {fold+1}: {e_eval_new}")
                    current_fold_new_data_recall = 0.0 # אפס את ה-recall אם ההערכה נכשלה
            else:
                print(f"    Fold {fold+1} best model (ValAcc={best_val_acc_fold:.4f}) is outside the threshold [{MIN_VAL_ACC_THRESHOLD}, {MAX_VAL_ACC_THRESHOLD}]. Skipping new data evaluation and model saving for this fold.")
        elif new_specs_np.size == 0:
            print(f"    Skipping new data evaluation for fold {fold+1} as no new data was loaded.")
        else: # אם current_best_fold_model_state הוא None
             print(f"    Skipping new data evaluation for fold {fold+1} as no suitable model was found (e.g., training failed).")


        # הוסף את התוצאות של ה-fold הנוכחי לרשימות הכלליות של ה-trial
        fold_recalls_new_data.append(current_fold_new_data_recall) # הוסף 0 אם ההערכה לא בוצעה או נכשלה
        fold_val_accs.append(best_val_acc_fold) # שמור את ה-val_acc הטוב ביותר של ה-fold בכל מקרה

        # --- 3.5 דיווח ל-Optuna ובדיקת Pruning ---
        # דווח על המדד העיקרי (למשל, Recall על נתונים חדשים) באמצע הדרך
        # זה מאפשר ל-Optuna להחליט אם להפסיק את ה-trial מוקדם (pruning)
        # <<< תיקון: הוספת try-except סביב פעולות Optuna >>>
        try:
            # דווח על ה-recall שהושג עד כה ב-fold זה
            # אם ההערכה לא בוצעה/נכשלה, הערך יהיה 0
            trial.report(current_fold_new_data_recall, fold)

            # בדוק אם Optuna מציע להפסיק את ה-trial
            if trial.should_prune():
                print(f"  Trial {trial.number} pruned at fold {fold+1}.")
                raise optuna.exceptions.TrialPruned() # זרוק חריגה כדי לצאת מה-trial

        except optuna.exceptions.TrialPruned:
             raise # זרוק מחדש את החריגה כדי ש-Optuna יתפוס אותה
        except Exception as e_optuna_report:
             print(f"  Warning: Optuna reporting/pruning failed for fold {fold+1}: {e_optuna_report}")
             # אל תזרוק שגיאה, רק הדפס ותמשיך את ה-trial

        print(f"  --- Fold {fold+1} Finished. Time: {time.time() - fold_start_time:.1f}s ---")


    # --- 4. חישוב תוצאה סופית ל-Trial ---
    # ממוצע של המדדים על פני כל ה-folds
    average_recall_new_data = np.mean(fold_recalls_new_data) if fold_recalls_new_data else 0.0
    average_val_acc = np.mean(fold_val_accs) if fold_val_accs else 0.0
    std_recall_new_data = np.std(fold_recalls_new_data) if len(fold_recalls_new_data) > 1 else 0.0
    std_val_acc = np.std(fold_val_accs) if len(fold_val_accs) > 1 else 0.0

    print(f"\n--- Finished Trial {trial.number} ---")
    print(f"  Avg Val Acc: {average_val_acc:.4f} (+/- {std_val_acc:.4f})")
    print(f"  Avg New Data Recall (Steps): {average_recall_new_data:.4f} (+/- {std_recall_new_data:.4f})")
    print(f"  Total Qualified Fold Models Saved in Memory: {len(fold_best_models_info)}")
    print(f"  Total Trial Time: {time.time() - start_trial_time:.1f}s")
    print(f"{'='*40}")

    # --- 5. שמירת מידע על Trial מוצלח ב-User Attributes ---
    # שמור את פרטי ה-trial הזה אם הוא נראה מבטיח (למשל, השיג recall חיובי כלשהו)
    # ורק אם נשמרו מודלים כלשהם מה-folds שלו
    if average_recall_new_data > 0 and fold_best_models_info:
        trial_summary = {
            'trial_num': trial.number,
            'params': trial.params,
            'avg_val_acc': average_val_acc,
            'std_val_acc': std_val_acc,
            'avg_recall_new': average_recall_new_data,
            'std_recall_new': std_recall_new_data,
            # שים לב: 'fold_models' עשוי להכיל הרבה state_dicts ולהיות גדול מאוד בזיכרון
            # עדיף לשמור רק את המדדים כאן, ולשמור את ה-state_dicts לקבצים בסוף
            'num_qualified_folds': len(fold_best_models_info),
            # אפשר לשמור כאן רק את המדדים של ה-folds, לא את ה-state_dicts
            'qualified_fold_metrics': [{'fold': f['fold'], 'val_acc': f['val_acc'], 'new_recall': f['new_recall']} for f in fold_best_models_info]
        }
        # שמור גם את המידע המלא (עם state_dicts) באופן זמני ב-trial עצמו
        # כדי שנוכל לגשת אליו מחוץ לפונקציית objective אם זה ה-trial הטוב ביותר
        trial.set_user_attr("full_trial_info_with_models", {
            'summary': trial_summary,
            'fold_models': fold_best_models_info # זה יכול להיות גדול!
        })

        # עדכן את ה-user_attr של ה-study כולו עם ה-trial הטוב ביותר שנמצא עד כה
        try:
            # קבל את ה-recall הטוב ביותר הנוכחי מה-study
            current_best_recall = study.user_attrs.get("best_overall_recall", -1.0)
            if average_recall_new_data > current_best_recall:
                print(f"  >>> Trial {trial.number} is the new best overall! Saving info to study attributes. <<<")
                study.set_user_attr("best_overall_recall", average_recall_new_data)
                # שמור רק את הסיכום ב-study attribute הראשי (ללא state_dicts)
                study.set_user_attr("best_trial_summary", trial_summary)
        except Exception as e_attr:
            print(f"  Warning: Could not set/get user_attrs for best trial tracking: {e_attr}")


    # --- 6. החזרת ערך המטרה ל-Optuna ---
    # החזר את המדד ש-Optuna צריך למקסם (או למזער)
    # במקרה זה, נמקסם את ממוצע ה-Recall על הנתונים החדשים
    return average_recall_new_data


# --- 5. הרצת החיפוש עם Optuna ---
print("\n--- Starting Optuna Hyperparameter Optimization ---")

study = None
try:
    # נסה לטעון study קיים
    study = joblib.load(STUDY_FILENAME)
    print(f"Loaded existing Optuna study '{STUDY_NAME_OPT}' from {STUDY_FILENAME}")
    # עדכן את best_overall_recall אם קיים ב-study הנטען
    if "best_overall_recall" not in study.user_attrs:
         study.set_user_attr("best_overall_recall", -1.0) # אתחול אם חסר

except FileNotFoundError:
    print(f"No existing study found at {STUDY_FILENAME}. Creating a new study.")
    # צור study חדש אם הקובץ לא נמצא
    # מגדירים למקסם את הערך המוחזר מ-objective (שהוא average_recall_new_data)
    study = optuna.create_study(study_name=STUDY_NAME_OPT, direction="maximize")
    study.set_user_attr("best_overall_recall", -1.0) # אתחול המדד הטוב ביותר
except Exception as e_load:
    print(f"Error loading study from {STUDY_FILENAME}: {e_load}. Creating a new one.")
    study = optuna.create_study(study_name=STUDY_NAME_OPT, direction="maximize")
    study.set_user_attr("best_overall_recall", -1.0)

# הגדרת Callback לשמירת ה-study אחרי כל trial (או כל מספר trials)
save_callback = lambda study, trial: joblib.dump(study, STUDY_FILENAME)
print(f"Study will be saved to {STUDY_FILENAME} after each trial.")

# הרצת האופטימיזציה
try:
    print(f"Starting optimization run for max {TIMEOUT_SECONDS_OPT} seconds...")
    study.optimize(objective,
                   # n_trials=50, # אפשר להגדיר מספר trials במקום זמן
                   timeout=TIMEOUT_SECONDS_OPT,
                   callbacks=[save_callback],
                   # gc_after_trial=True # נסה להפעיל פינוי זיכרון אם יש בעיות
                   )
except KeyboardInterrupt:
    print("\nOptimization interrupted by user. Saving current study state.")
    joblib.dump(study, STUDY_FILENAME)
except Exception as e_optimize:
    print(f"\nAn error occurred during optimization: {e_optimize}")
    print("Attempting to save current study state.")
    try:
        joblib.dump(study, STUDY_FILENAME)
    except Exception as e_save_fail:
        print(f"Could not save study after error: {e_save_fail}")

# --- 6. סיכום, שמירת Top N מודלים, והצגת דוח ---
print("\n--- Optimization Finished ---")

# הדפסת סיכום התוצאות מ-Optuna
print(f"Number of finished trials: {len(study.trials)}")

# קבל את ה-trial הטוב ביותר לפי הערך המוחזר (average_recall_new_data)
best_trial_optuna = None
try:
    best_trial_optuna = study.best_trial
    print(f"\nBest trial found by Optuna (based on return value):")
    print(f"  Trial Number: {best_trial_optuna.number}")
    print(f"  Value (Avg New Recall): {best_trial_optuna.value:.4f}")
    print(f"  Params: {best_trial_optuna.params}")
except ValueError:
    print("\nNo completed trials found in the study.")
except Exception as e_best:
    print(f"Could not retrieve best trial from Optuna: {e_best}")

# קבל את המידע על ה-trial הטוב ביותר ששמרנו ב-user_attrs (יכול להיות שונה אם היה שגיאה בשמירה)
best_trial_summary_attr = study.user_attrs.get("best_trial_summary", None)
if best_trial_summary_attr:
    print("\nBest trial summary stored in study user attributes:")
    print(f"  Trial Number: {best_trial_summary_attr.get('trial_num', 'N/A')}")
    print(f"  Avg Val Acc: {best_trial_summary_attr.get('avg_val_acc', -1):.4f}")
    print(f"  Avg New Recall: {best_trial_summary_attr.get('avg_recall_new', -1):.4f}")
    print(f"  Params: {best_trial_summary_attr.get('params', {})}")
    print(f"  Num Qualified Folds: {best_trial_summary_attr.get('num_qualified_folds', 0)}")
else:
    print("\nNo best trial summary found in study user attributes.")


# --- שמירת המודלים מה-Top N Trials ---
print(f"\n--- Saving Top {TOP_N_MODELS} Models ---")

# אסוף את כל ה-trials שהיו מוצלחים (השיגו recall > 0 ויש להם fold_models)
successful_trials_info = []
for trial in study.trials:
    if trial.state == optuna.trial.TrialState.COMPLETE and trial.value is not None and trial.value > 0:
        # נסה לקבל את המידע המלא מה-user_attr של ה-trial
        full_info = trial.user_attrs.get("full_trial_info_with_models", None)
        if full_info and 'fold_models' in full_info and full_info['fold_models']:
            successful_trials_info.append(full_info)

if not successful_trials_info:
    print("No successful trials with saved models found to select from.")
else:
    # מיין את ה-trials המוצלחים לפי ה-recall הממוצע (מהגבוה לנמוך)
    # השתמש בערך מ-summary אם קיים, אחרת בערך שהוחזר מה-trial
    sorted_trials = sorted(successful_trials_info,
                           key=lambda x: x['summary'].get('avg_recall_new', -1),
                           reverse=True)

    # קח את ה-Top N trials
    top_n_trials = sorted_trials[:TOP_N_MODELS]
    print(f"Found {len(successful_trials_info)} successful trials. Selecting top {len(top_n_trials)}.")

    # שמור את ה-state_dict של המודלים מכל fold שעמד בתנאים ב-trials הטובים ביותר
    saved_model_count = 0
    for rank, trial_info in enumerate(top_n_trials):
        trial_num = trial_info['summary']['trial_num']
        avg_recall = trial_info['summary']['avg_recall_new']
        print(f"\nSaving models for Rank {rank+1} Trial {trial_num} (Avg Recall: {avg_recall:.4f})")

        if not trial_info.get('fold_models'):
            print("  No fold models found in this trial's info.")
            continue

        for fold_model_info in trial_info['fold_models']:
            fold_num = fold_model_info['fold']
            val_acc = fold_model_info['val_acc']
            new_recall = fold_model_info['new_recall']
            epoch = fold_model_info['epoch']
            state_dict = fold_model_info['state_dict']

            # הגדר שם קובץ ברור
            filename = f"trial_{trial_num}_rank_{rank+1}_fold_{fold_num}_epoch_{epoch}_valacc_{val_acc:.3f}_newrecall_{new_recall:.3f}.pth"
            save_path = os.path.join(SAVE_PATH_BASE, filename)

            try:
                torch.save(state_dict, save_path)
                print(f"  Saved: {filename}")
                saved_model_count += 1

                # --- אופציונלי: שרטט גרפים עבור המודלים השמורים ---
                plot_fold_results(
                    fold_num=fold_num,
                    history=fold_model_info.get('history'),
                    val_labels=fold_model_info.get('val_labels'),
                    val_preds=fold_model_info.get('val_preds'),
                    new_data_labels=fold_model_info.get('new_labels'),
                    new_data_preds=fold_model_info.get('new_preds'),
                    new_data_recall=new_recall,
                    class_names=CLASS_NAMES,
                    save_path=SAVE_PATH_BASE, # שמור גרפים באותה תיקייה
                    trial_num=trial_num
                )

            except Exception as e_save_plot:
                print(f"  ERROR saving model or plotting for {filename}: {e_save_plot}")

    print(f"\nSuccessfully saved {saved_model_count} model state dictionaries to {SAVE_PATH_BASE}")

# --- הצגת גרפים וסיכומים נוספים של Optuna (אופציונלי) ---
if len(study.trials) > 1:
    try:
        # גרף היסטוריית האופטימיזציה
        fig_hist = optuna.visualization.plot_optimization_history(study)
        fig_hist.show()
        # שמירת הגרף לקובץ
        hist_path = os.path.join(SAVE_PATH_BASE, f"{STUDY_NAME_OPT}_optimization_history.png")
        fig_hist.write_image(hist_path)
        print(f"Saved optimization history plot to {hist_path}")


        # גרף חשיבות פרמטרים (דורש מספר trials)
        fig_importance = optuna.visualization.plot_param_importances(study)
        fig_importance.show()
        importance_path = os.path.join(SAVE_PATH_BASE, f"{STUDY_NAME_OPT}_param_importances.png")
        fig_importance.write_image(importance_path)
        print(f"Saved parameter importances plot to {importance_path}")


        # גרף קואורדינטות מקבילות
        fig_coord = optuna.visualization.plot_parallel_coordinate(study)
        fig_coord.show()
        coord_path = os.path.join(SAVE_PATH_BASE, f"{STUDY_NAME_OPT}_parallel_coordinate.png")
        fig_coord.write_image(coord_path)
        print(f"Saved parallel coordinate plot to {coord_path}")

    except (ImportError, ModuleNotFoundError):
        print("\nInstall plotly and kaleido (pip install plotly kaleido) to generate Optuna visualization plots.")
    except Exception as e_viz:
        print(f"\nCould not generate Optuna visualization plots: {e_viz}")


print("\n--- Script Finished ---")
```
